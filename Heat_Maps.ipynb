```python
import numpy as np # library to handle data in a vectorized manner

import pandas as pd # library for data analsysis
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

import json # library to handle JSON files

!conda install -c conda-forge geopy --yes # uncomment this line if you haven't completed the Foursquare API lab
from geopy.geocoders import Nominatim # convert an address into latitude and longitude values

import requests # library to handle requests
from pandas.io.json import json_normalize # tranform JSON file into a pandas dataframe

# Matplotlib and associated plotting modules
import matplotlib.cm as cm
import matplotlib.colors as colors

# import k-means from clustering stage
from sklearn.cluster import KMeans

#!conda install -c conda-forge folium=0.5.0 --yes # uncomment this line if you haven't completed the Foursquare API lab
import folium # map rendering library
from folium import plugins
from folium.plugins import HeatMap

print('Libraries imported.')
```

```python
df = pd.read_excel('./traffic-signal-vehicle-and-pedestrian-volumes-data.xlsx')
df.describe()
```

```python
# create map of Toronto using latitude and longitude values
map_veh_toronto = folium.Map(location=[43.6, -79.2], zoom_start=10)
max_vehicle_vol = df['8 Peak Hr Vehicle Volume'].max()
vehicle_map = list(zip(df['Latitude'], df['Longitude'], df['8 Peak Hr Vehicle Volume']/max_vehicle_vol))
plugins.HeatMap(vehicle_map, radius=5, blur=1).add_to(map_veh_toronto)
map_veh_toronto
```

```python
map_ped_toronto = folium.Map(location=[43.6, -79.2], zoom_start=10)
max_pedestrian_vol = df['8 Peak Hr Pedestrian Volume'].max()
ped_map = list(zip(df['Latitude'], df['Longitude'], df['8 Peak Hr Pedestrian Volume']/max_pedestrian_vol))
plugins.HeatMap(ped_map, radius=5, blur=1).add_to(map_ped_toronto)
map_ped_toronto
```
